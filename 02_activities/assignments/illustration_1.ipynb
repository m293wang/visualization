```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
file_path = r'C:\Users\m293w\Documents\Temp\visualization\assignment-3\ontario_top_baby_names_female.csv'
data = pd.read_csv(file_path)
data.head() #check data
```

```python
#Plot 1
data_2023 = data[data['Year/Année'] == 2023] #select only 2023 data
top_20 = data_2023.sort_values('Frequency/Fréquence', ascending=False).head(20) #select top 20 from 2023
x_top20 = top_20['Name/Nom']
y_top20 = top_20['Frequency/Fréquence']

fig, ax = plt.subplots(figsize = (8, 4))
bars = ax.bar(x_top20, y_top20, color = 'pink')
ax.set_title('Top 20 female baby names in Ontario 2023')
ax.set_ylabel('Frequency')
ax.set_xlabel('Name')
ax.set_xticklabels(x_top20, rotation=90)
ax.bar_label(bars, fontsize = 6)
```

```python
#plot 2
data_m = data[data['Name/Nom']=='MINA'] #select all 'MINA's
x_m = data_m['Year/Année']
y_m = data_m['Frequency/Fréquence']
fig, ax = plt.subplots(figsize = (10, 3))
ax.plot(x_m, y_m,
        marker = '*',
        color = 'pink',
        linestyle = '--',
        linewidth = 2,
        markersize = 3,
        markeredgecolor = 'red',
        markerfacecolor = 'pink')
ax.set_title('Frequency of "Mina" as a baby girl name in Ontario through history')
ax.set_xlabel('Year')
ax.set_ylabel('Frequency')
```
