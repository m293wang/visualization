```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize = (5,3))
#ax.scatter(x, y)
#ax.bar(x, y)
#ax.plot(x, y)
ax.hist(y)
```

```python
font1 = {'family':'sans-serif', 'color': 'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total grwoth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y,
           marker = '*',
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y,
        marker = '*',
        color = 'indigo',
        linestyle = '--',
        linewidth = 2,
        markersize = 12,
        markeredgecolor = 'yellow',
        markerfacecolor = 'black')

ax.grid(axis = 'y',
        color = 'blue',
        linewidth = 1,
        linestyle = 'dotted')
```
